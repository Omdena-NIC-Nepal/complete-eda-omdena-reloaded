## Introduction
This notebook contains an exploratory data analysis of climate data from 1900 to 2023. The dataset includes global temperatures, CO2 concentration, sea level rise, and Arctic ice area.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styling
plt.style.use('seaborn-whitegrid')
sns.set_palette('viridis')
%matplotlib inline
```

## 1. Data Preparation

Load the climate data and perform necessary cleaning and aggregation.

```python
# Load the dataset
df = pd.read_csv('./data/Climate_Change_Indicators.csv') # Place the correct path to the file you are reading here (Make sure to load using the relative path)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values and basic information about the dataset
print("Dataset Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# TODO: Aggregate data by year to create a 124-year time series
# Your code here
yearly_average_data = df.groupby(by="Year").agg('mean')
yearly_average_data.reset_index(inplace=True)
yearly_average_data.head()
```

## 2. Univariate Analysis

Analyze each climate variable independently.

```python
# TODO: Perform univariate analysis for each climate variable
# Include descriptive statistics and appropriate visualizations
# Your code here
print(yearly_average_data[['Global Average Temperature (°C)','CO2 Concentration (ppm)',
                           'Sea Level Rise (mm)','Arctic Ice Area (million km²)']].describe())
```

```python
yearly_average_data.head()
```

```python
sns.lineplot(x='Year', y='Global Average Temperature (°C)', data=yearly_average_data)
plt.title('Yearly average Global Average Temperature (°C)')
plt.show()
```

```python
sns.histplot(x='CO2 Concentration (ppm)', data=yearly_average_data, bins=20, kde=True)
plt.title('Histogram of yearly average CO2 Concentration (ppm)')
plt.show()
```

```python
sns.scatterplot(x='Year', y='Arctic Ice Area (million km²)', data=yearly_average_data)
plt.title('Scatter plot of yearly average Arctic Ice Area (million km²)')
plt.show()
```

```python
yearly_average_data.columns
```

```python
sns.boxplot(y='Sea Level Rise (mm)', data=yearly_average_data)
plt.title('Scatter plot of yearly average Sea Level Rise (mm)')
plt.show()
```

## 3. Bivariate Analysis

Explore relationships between pairs of climate variables.

```python
yearly_average_data.columns
```

```python
# TODO: Perform bivariate analysis
# Include correlation analysis and appropriate visualizations
# Your code here
sns.regplot(x='Sea Level Rise (mm)', y='Arctic Ice Area (million km²)', data=yearly_average_data, )
plt.title('Scatter plot of "Sea Level Rise (mm)" vs "Arctic Ice Area (million km²)"')
plt.show()
```

```python
sns.pairplot(data=yearly_average_data[['Global Average Temperature (°C)', 'CO2 Concentration (ppm)',
       'Sea Level Rise (mm)', 'Arctic Ice Area (million km²)']], kind='reg')
plt.title('Pair plot')
plt.show()
```

```python
corr_matrix = yearly_average_data[['Global Average Temperature (°C)', 'CO2 Concentration (ppm)',
       'Sea Level Rise (mm)', 'Arctic Ice Area (million km²)']].corr()
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.title('Heatmap plot')
plt.show()
```
