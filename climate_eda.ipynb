```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the climate change dataset
df = pd.read_csv('Climate_Change_Indicators.csv')

# Display the first few rows to inspect the data
df.head()

# Data Cleaning: Handle missing values if any (optional)
df = df.dropna()  # Drops rows with any missing values
# Or fill missing values (for specific columns, if necessary)
# df['CO2 Concentration (ppm)'] = df['CO2 Concentration (ppm)'].fillna(df['CO2 Concentration (ppm)'].mean())

# Ensure the 'Year' column is of type int
df['Year'] = df['Year'].astype(int)

# Yearly aggregation
df_yearly = df.groupby('Year').agg({
    'Global Average Temperature (°C)': 'mean',
    'CO2 Concentration (ppm)': 'mean',
    'Sea Level Rise (mm)': 'mean',
    'Arctic Ice Area (million km²)': 'mean'
}).reset_index()

# Display the aggregated yearly data
df_yearly.head()
```

```python
# Univariate Analysis: Visualizations

# Histogram for Global Average Temperature
plt.figure(figsize=(8, 6))
sns.histplot(df['Global Average Temperature (°C)'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Global Average Temperature Distribution (°C)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()

# Boxplot for CO2 Concentration
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['CO2 Concentration (ppm)'], color='salmon')
plt.title('CO2 Concentration Distribution (ppm)')
plt.xlabel('CO2 Concentration (ppm)')
plt.show()

# Line plot for Sea Level Rise over time
plt.figure(figsize=(8, 6))
sns.lineplot(x='Year', y='Sea Level Rise (mm)', data=df, color='green')
plt.title('Sea Level Rise Over Time (mm)')
plt.xlabel('Year')
plt.ylabel('Sea Level Rise (mm)')
plt.show()

# Boxplot for Arctic Ice Area
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Arctic Ice Area (million km²)'], color='lightgreen')
plt.title('Arctic Ice Area Distribution (million km²)')
plt.xlabel('Arctic Ice Area (million km²)')
plt.show()

# Descriptive statistics
df.describe()
```

```python
# Bivariate Analysis: Relationships between pairs of variables

# Scatter plot: Global Average Temperature vs CO2 Concentration
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Global Average Temperature (°C)', y='CO2 Concentration (ppm)', data=df)
plt.title('Global Average Temperature vs CO2 Concentration')
plt.xlabel('Global Average Temperature (°C)')
plt.ylabel('CO2 Concentration (ppm)')
plt.show()

# Heatmap of correlations between variables
plt.figure(figsize=(8, 6))
corr_matrix = df[['Global Average Temperature (°C)', 'CO2 Concentration (ppm)', 'Sea Level Rise (mm)', 'Arctic Ice Area (million km²)']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Multivariate Analysis: Visualizations

# Pairplot for multiple climate variables
sns.pairplot(df[['Global Average Temperature (°C)', 'CO2 Concentration (ppm)', 'Sea Level Rise (mm)', 'Arctic Ice Area (million km²)']])
plt.suptitle('Pairplot of Climate Variables', y=1.02)
plt.show()
```

## Conclusion and Insights

1. **Global Average Temperature** has shown a steady increase, particularly in recent decades, indicating significant global warming.
2. **CO2 Concentration** is strongly correlated with global temperature rise, reinforcing the impact of greenhouse gas emissions on climate change.
3. **Sea Level Rise** has been gradual but continuous, highlighting the effects of melting polar ice and thermal expansion due to rising temperatures.
4. **Arctic Ice Area** has been consistently declining, which is a clear indicator of the ongoing warming in the Arctic region and its environmental consequences.
5. Based on these findings, urgent action is needed to mitigate the effects of climate change and reduce CO2 emissions to prevent further global warming and environmental damage.
