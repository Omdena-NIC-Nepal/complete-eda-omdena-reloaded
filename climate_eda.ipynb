## Introduction
This notebook contains an exploratory data analysis of climate data from 1900 to 2023. The dataset includes global temperatures, CO2 concentration, sea level rise, and Arctic ice area.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styling
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')
%matplotlib inline
```

## 1. Data Preparation

Load the climate data and perform necessary cleaning and aggregation.

```python
# Load the dataset
df = pd.read_csv("C:/Users/Administrator/OneDrive/Desktop/complete-eda-omdena-reloaded/data/Climate_Change_Indicators.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values and basic information about the dataset
print("Dataset Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# TODO: Aggregate data by year to create a 124-year time series
df['Year'] = pd.to_datetime(df['Year'], format='%Y')  # Convert 'Year' to datetime
df_yearly = df.groupby(df['Year'].dt.year).mean()  # Group by year and calculate the mean for each variable

# Display the aggregated data
df_yearly.head()
```

## 2. Univariate Analysis

Analyze each climate variable independently.

```python
# TODO: Perform univariate analysis for each climate variable
# Include descriptive statistics and appropriate visualizations

# Univariate analysis for each variable

# Descriptive statistics
print("Descriptive Statistics:")
print(df_yearly.describe())

# Global Temperature
plt.figure(figsize=(10, 5))
sns.histplot(df_yearly['Global Average Temperature (°C)'], kde=True, bins=20)
plt.title('Global Average Temperature Distribution')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df_yearly['Global Average Temperature (°C)'])
plt.title('Global Average Temperature Boxplot')
plt.show()

# CO2 Concentration
plt.figure(figsize=(10, 5))
sns.histplot(df_yearly['CO2 Concentration (ppm)'], kde=True, bins=20)
plt.title('CO2 Concentration Distribution')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df_yearly['CO2 Concentration (ppm)'])
plt.title('CO2 Concentration Boxplot')
plt.show()

# Sea Level Rise
plt.figure(figsize=(10, 5))
sns.histplot(df_yearly['Sea Level Rise (mm)'], kde=True, bins=20)
plt.title('Sea Level Rise Distribution')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df_yearly['Sea Level Rise (mm)'])
plt.title('Sea Level Rise Boxplot')
plt.show()

# Arctic Ice Area
plt.figure(figsize=(10, 5))
sns.histplot(df_yearly['Arctic Ice Area (million km²)'], kde=True, bins=20)
plt.title('Arctic Ice Area Distribution')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df_yearly['Arctic Ice Area (million km²)'])
plt.title('Arctic Ice Area Boxplot')
plt.show()
```

## 3. Bivariate Analysis

Explore relationships between pairs of climate variables.

```python
# TODO: Perform bivariate analysis
# Include correlation analysis and appropriate visualizations

# Bivariate analysis for relationships between variables

# Scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_yearly['Global Average Temperature (°C)'], y=df_yearly['CO2 Concentration (ppm)'])
plt.title('Global Temperature vs CO2 Concentration')
plt.xlabel('Global Average Temperature (°C)')
plt.ylabel('CO2 Concentration (ppm)')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_yearly['Sea Level Rise (mm)'], y=df_yearly['Arctic Ice Area (million km²)'])
plt.title('Sea Level Rise vs Arctic Ice Area')
plt.xlabel('Sea Level Rise (mm)')
plt.ylabel('Arctic Ice Area (million km²)')
plt.show()

# Correlation analysis
corr_matrix = df_yearly.corr()

# Heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

## 4. Multivariate Analysis

Investigate relationships among three or more variables.

```python
# TODO: Perform multivariate analysis
# Create advanced visualizations showing multiple variables

# Multivariate analysis for relationships among three or more variables

# Pairplot for multivariate visualization
sns.pairplot(df_yearly[['Global Average Temperature (°C)', 'CO2 Concentration (ppm)', 
                        'Sea Level Rise (mm)', 'Arctic Ice Area (million km²)']])
plt.suptitle('Pairplot of Climate Variables', y=1.02)
plt.show()

# 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_yearly['Global Average Temperature (°C)'], 
           df_yearly['CO2 Concentration (ppm)'], 
           df_yearly['Sea Level Rise (mm)'], c='blue')
ax.set_xlabel('Global Average Temperature (°C)')
ax.set_ylabel('CO2 Concentration (ppm)')
ax.set_zlabel('Sea Level Rise (mm)')
plt.title('3D Scatter Plot of Climate Variables')
plt.show()
```
